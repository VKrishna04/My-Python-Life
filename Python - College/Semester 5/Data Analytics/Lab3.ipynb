Pandas Series

```python
import pandas as pd
import numpy as np

# Creating empty series
ser = pd.Series()
display(f"Pandas Series: \n{ser}")

# Simple Array
data = np.array([_ for _ in "geeks"])

# Creating filled series
ser = pd.Series(data)
display(f"Pandas Series: \n{ser}")
```

```python
import pandas as pd

series = [_ for _ in "CSE"]
ser = pd.Series(series)
display(ser)
```

Data Frames

```python
import pandas as pd

data = {
    "Name": ["Tom", "will", "kill", "pandey"],
    "Age": [
        20,
        21,
        22,
        23,
    ],
}
df = pd.DataFrame(data)
display(df)
```

```python
import pandas as pd

data = {
    "Name": ["Jai", "Princi", "Gaurav", "Anuj"],
    "Age": [27, 24, 22, 32],
    "Adress": ["Delhi", "Kanpur", "Allahabad", "Kannauj"],
    "Qualification": ["Msc", "MA", "MCA", "Phd"],
}
df = pd.DataFrame(data)
display(df)
display(df[['Name','Qualification']])
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv(r'.\Datasets\nba.csv')

# Display location
print(df.iloc[0])
print(df.iloc[0:0])

# Display the first 5 rows of the DataFrame
print("Head of the DataFrame:")
display(df.head())

# Display the last 5 rows of the DataFrame
print("\nTail of the DataFrame:")
display(df.tail())

# Display the info of the DataFrame
print("\nInfo of the DataFrame:")
display(df.info())

# Display the descriptive statistics of the DataFrame
print("\nDescription of the DataFrame:")
display(df.describe())
```
