```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the first line
x1 = np.array([1, 2, 3, 4])
y1 = x1 ** 2

# Plotting the first line with solid line and circular markers
plt.plot(x1, y1, linestyle='-', marker='o', color='green', linewidth=2, markersize=8, label='x squared')

# Data for the second line
x2 = np.array([1, 2, 3, 4])
y2 = x2 ** 3

# Plotting the second line with dotted line and square markers
plt.plot(x2, y2, linestyle=':', marker='s', color='purple', linewidth=2, markersize=8, label='x cubed')

# Adding labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Customized Line Graph')

# Adding a legend
plt.legend(loc='best')

# Adding gridlines
plt.grid(True, linestyle='--', alpha=0.6)

# Annotating specific points
for i in range(len(x1)):
    plt.annotate(f'({x1[i]}, {y1[i]})', (x1[i], y1[i]), textcoords="offset points", xytext=(0,5), ha='center', fontsize=8, color='black')
for i in range(len(x2)):
    plt.annotate(f'({x2[i]}, {y2[i]})', (x2[i], y2[i]), textcoords="offset points", xytext=(0,5), ha='center', fontsize=8, color='black')

# Displaying the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating data
x = np.linspace(-10, 10, 100)
y = np.sin(x)

# Plotting the sine wave with dashed line style and circular markers
plt.plot(x, y, linestyle='--', marker='o', color='magenta', markersize=5, linewidth=2, label='sin(x)')

# Filling the area under the sine wave with a light color
plt.fill_between(x, y, color='magenta', alpha=0.1)

# Adding labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sine Wave with Fill and Markers')

# Adding a legend
plt.legend(loc='best')

# Adding gridlines
plt.grid(True, linestyle='--', alpha=0.6)

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating data
x = np.linspace(-10, 10, 100)

# Creating a grid of plots with two rows
fig, ax = plt.subplots(2, figsize=(8, 6))  # Added figsize for better layout

# Plotting sine wave in the first subplot
ax[0].plot(x, np.sin(x), linestyle='-', color='blue', label='sin(x)')
ax[0].set_title('Sine Wave')
ax[0].set_xlabel('x-axis')
ax[0].set_ylabel('y-axis')
ax[0].legend()
ax[0].grid(True, linestyle='--', alpha=0.6)

# Plotting cosine wave in the second subplot
ax[1].plot(x, np.cos(x), linestyle='--', color='red', label='cos(x)')
ax[1].set_title('Cosine Wave')
ax[1].set_xlabel('x-axis')
ax[1].set_ylabel('y-axis')
ax[1].legend()
ax[1].grid(True, linestyle='--', alpha=0.6)

# Adjusting layout to fit plots nicely
plt.tight_layout()

# Displaying the plots
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values = [15, 24, 30]

# Create a horizontal bar graph
plt.barh(categories, values, color='skyblue')

# Labeling the axes correctly for a horizontal bar graph
plt.xlabel('Values')
plt.ylabel('Categories')

# Title updated to reflect the graph type
plt.title('Horizontal Bar Graph')

plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['Category A', 'Category B', 'Category C']
values = [15, 24, 30]

plt.barh(categories, values, color=['green', 'orange', 'blue'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Vertical Bar Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Defining categories and their corresponding values for two groups
categories = ['Category A', 'Category B', 'Category C']
values1 = [15, 24, 30]
values2 = [20, 18, 25]

# Setting the width of the bars
bar_width = 0.35
# Calculating bar positions for both groups
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width

# Creating the first set of bars (Group 1)
plt.bar(bar_positions1, values1, width=bar_width, label='Group 1', color='skyblue')
# Create the second set of bars (Group 2) next to the first set
plt.bar(bar_positions2, values2, width=bar_width, label='Group 2', color='orange')

# Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')

# Adding a title to the graph
plt.title('Grouped Bar Graph')

# Displaying a legend to identify the groups
plt.legend(loc='upper left')

# Adjusting the x-axis ticks and labels
plt.xticks(bar_positions1,categories,rotation=45)

# Showing the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
# Defining categories and values for two groups
categories = ['Category A', 'Category B', 'Category C']
values1 = [15, 24, 30]
values2 = [20, 18, 25]

# Creating the first set of bars (Group 1) without any offset
plt.bar(categories, values1, label='Group 1', color='skyblue')

# Creating the second set of bars (Group 2) plotted with 'bottom' set to the values of Group 1
# This makes Group 2 bars stacked on top of Group 1 bars
plt.bar(categories, values2, bottom=values1, label='Group 2', color='orange')

# Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')

plt.title('Stacked Bar Graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('movies_by_year.csv')

# Check if the 'Total Gross' column exists in the DataFrame
if 'Total Gross' not in df.columns:
    raise ValueError("The 'Total Gross' column is not found in the dataset.")

# Plot a histogram of 'Total Gross'
plt.figure(figsize=(10, 6))
plt.hist(df['Total Gross'], bins=30, color='blue', edgecolor='green')

# Customize the plot
plt.xlabel('Total Gross')
plt.ylabel('Frequency')
plt.title('Histogram of Total Gross')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('movies_by_year.csv')

# Check if the 'Total Gross' column exists in the DataFrame
if 'Total Gross' not in df.columns:
    raise ValueError("The 'Total Gross' column is not found in the dataset.")

# Plot a histogram of 'Total Gross' with customizable bins
bins = int(input("Enter the number of bins for the histogram: "))  # Allow user to specify the number of bins

plt.figure(figsize=(10, 6))
plt.hist(df['Total Gross'], bins=bins, color='skyblue', edgecolor='black')

# Customize the plot
plt.xlabel('Total Gross')
plt.ylabel('Frequency')
plt.title(f'Histogram of Total Gross with {bins} Bins')

# Display the plot
plt.show()
```

```python
#unequal bins
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('movies_by_year.csv')

# Check if the 'Total Gross' column exists in the DataFrame
if 'Total Gross' not in df.columns:
    raise ValueError("The 'Total Gross' column is not found in the dataset.")

# Ask the user for the number of unequal bins
num_bins = int(input("Enter the number of unequal bins: "))

# Generate unequal bin edges based on the 'Total Gross' range
min_value = df['Total Gross'].min()
max_value = df['Total Gross'].max()

# Creating unequal bins using logarithmic spacing for better visual distribution
bin_edges = np.geomspace(min_value + 1, max_value, num=num_bins + 1)  # Add 1 to include the endpoint

# Plot a histogram of 'Total Gross' with unequal bins
plt.figure(figsize=(10, 6))
plt.hist(df['Total Gross'], bins=bin_edges, color='skyblue', edgecolor='black')

# Customize the plot
plt.xscale('log')  # Set the x-axis to a logarithmic scale for better visualization
plt.xlabel('Total Gross (Log Scale)')
plt.ylabel('Frequency')
plt.title(f'Histogram of Total Gross (with {num_bins} unequal bins)')

# Display the plot
plt.show()
```
