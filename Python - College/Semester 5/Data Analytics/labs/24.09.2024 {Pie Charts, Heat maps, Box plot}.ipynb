```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
top_movies = pd.read_csv('top_movies.csv')

corr_matrix = top_movies[['Gross', 'Gross (Adjusted)']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Gross Earnings')
plt.show()
```

```python
df = pd.read_csv("top_movies.csv")

title_distribution = df.groupby('Studio').size().reset_index(name='count')
title_distribution = title_distribution.sort_values(by='count')

plt.figure(figsize=(15, 15))
plt.pie(
    title_distribution['count'], 
    labels=title_distribution['Studio'], 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Pastel1.colors
)
plt.title("Pie Chart of Movie Count by Studio")
plt.axis('equal')
plt.show()
```

```python
year_distribution = df.groupby('Year').size().reset_index(name='count')
year_distribution = year_distribution.sort_values(by='count')

plt.figure(figsize=(20, 20))
plt.pie(
    year_distribution['count'], 
    labels=year_distribution['Year'].astype(str), 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Pastel2.colors  
)
plt.title("Pie Chart of Number of Movies per Year")
plt.axis('equal')  
plt.show()
```
