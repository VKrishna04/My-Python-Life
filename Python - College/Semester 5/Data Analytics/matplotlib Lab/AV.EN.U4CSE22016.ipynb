# Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate random data for the histogram
i = 10000

data = np.random.randn(i)
print(f"Random {i} numbers generated")
```

```python
# Creating a customized histogram with a density plot
sns.histplot(
    data, bins=30, kde=True, color="teal", edgecolor="turquoise", linewidth=1.5
)

# Adding labels and title
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Customized Histogram with Density Plot")
```

```python
# Display the plot
plt.show()
```

Multiple Lines Charts

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x**2
plt.plot(x, y, "--", linewidth=3, color="black")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("First Chart")
# plt.grid(True)
plt.show()
plt.figure()
x = np.array([1, 2, 3, 4])

plt.plot(x, x**3, "-.", c="black", linewidth=3)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.title("Second Chart")
# plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x**2
plt.plot(x, y, "--", linewidth=3, color="black")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

x = np.array([1, 2, 3, 4])
plt.plot(x, x**3, "-.", linewidth=3, color="grey")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Line graph")
plt.legend(["first", "second"], loc="lower right")
# plt.grid(True)
plt.show()
```

```python
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.sin(x)
plt.plot(x, y, c="black")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sine Wave")
plt.legend(["sin(x)"])

plt.show()
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x), c="black")
ax[1].plot(x, np.cos(x), c="black")
```

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1)  # (rows, columns, panel number)
plt.plot(x, np.sin(x), c="black")

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), c="black")
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()  # create a plot figure
x = np.linspace(-10, 10, 100)
# create the first of two panels and set current axis
plt.subplot(2, 2, 1)  # (rows, columns, panel number)
plt.plot(x, np.sin(x), c="black")
plt.xlabel("x-values")
plt.ylabel("sine")


# create the second panel and set current axis
plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), c="black")
plt.xlabel("x-values")
plt.ylabel("cosine")

# create the third panel and set current axis
plt.subplot(2, 2, 3)
plt.plot(x, np.tan(x), c="black")
plt.xlabel("x-values")
plt.ylabel("tan")

# create the fourth panel and set current axis
plt.subplot(2, 2, 4)
plt.plot(x, x**2, c="black")
plt.xlabel("x-values")
plt.ylabel("x**2")
plt.tight_layout()
plt.show()
```

Grouped Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Defining categories and their corresponding values for two groups
categories = ["Category A", "Category B", "Category C"]
values1 = [15, 24, 30]
values2 = [20, 18, 25]

# Setting the width of the bars
bar_width = 0.35
# Calculating bar positions for both groups
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width

# Creating the first set of bars (Group 1)
plt.bar(bar_positions1, values1, width=bar_width, label="Group 1", color="lightgreen")
# Create the second set of bars (Group 2) next to the first set
plt.bar(bar_positions2, values2, width=bar_width, label="Group 2", color="lightblue")

# Adding labels to the axes
plt.xlabel("Categories")
plt.ylabel("Values")

# Adding a title to the graph
plt.title("Grouped Bar Graph")

# Displaying a legend to identify the groups
plt.legend(loc="upper left")

# Adjusting the x-axis ticks and labels
plt.xticks(bar_positions1, categories, rotation=45)

# Showing the plot
plt.show()
```

Stacked Bar Graph

```python
import matplotlib.pyplot as plt

# Defining categories and values for two groups
categories = ["Category A", "Category B", "Category C"]
values1 = [15, 24, 30]
values2 = [20, 18, 25]

# Creating the first set of bars (Group 1) without any offset
plt.bar(categories, values1, label="Group 1", color="lightblue")

# Creating the second set of bars (Group 2) plotted with 'bottom' set to the values of Group 1
# This makes Group 2 bars stacked on top of Group 1 bars
plt.bar(categories, values2, bottom=values1, label="Group 2", color="lightgreen")

# Adding labels to the axes
plt.xlabel("Categories")
plt.ylabel("Values")

plt.title("Stacked Bar Graph")
plt.legend()
plt.show()
```
