### Task 3: Handling Missing Data and Outliers

#### Explanation:
Missing data is handled by filling numerical columns with their mean values. To detect and remove outliers, we use the Z-score method, which identifies values that are more than three standard deviations from the mean. The dataset is then filtered to remove these outliers.

#### Code:

```python
# Fill missing values with mean for numerical columns
df.fillna(df.select_dtypes(include=["float64", "int64"]).mean(), inplace=True)

# Detect outliers using Z-score
from scipy.stats import zscore

# Calculate Z-scores and filter out data points with high Z-scores
z_scores = zscore(df.select_dtypes(include=["float64", "int64"]))
df_cleaned = df[(z_scores < 3).all(axis=1)]

print(df_cleaned.shape)
```

### Task 4: Converting Categorical Data to Numerical

#### Explanation:
Categorical columns are converted to numerical format using one-hot encoding. This process transforms categorical variables into a format suitable for machine learning algorithms by creating new binary columns for each category.

#### Code:

```python
# Convert categorical columns using one-hot encoding
df_encoded = pd.get_dummies(df_cleaned, drop_first=True)

display(df_encoded.head())
```

#### **2. Inferential Analysis**

##### **Hypothesis Testing**
**Theory:**
Hypothesis testing is used to make conclusions about a population.
- **Null Hypothesis (\(H_0\))**: Assumes no effect or difference exists.
- **Alternative Hypothesis (\(H_a\))**: Suggests an effect or difference exists.

**Code:**

```python
from scipy.stats import ttest_ind

# Example Hypothesis: Test if 'population_total' differs between two states
group1 = df[df["state_name"] == "PUNJAB"]["population_total"]
group2 = df[df["state_name"] == "MAHARASHTRA"]["population_total"]

# Perform t-test
t_stat, p_val = ttest_ind(group1, group2, nan_policy="omit")

print("T-Statistic:", t_stat)
print("P-Value:", p_val)

if p_val < 0.05:
    print("Reject the Null Hypothesis: Significant difference exists.")
else:
    print("Fail to Reject the Null Hypothesis: No significant difference.")
```

##### **Bootstrapping**
**Theory:**
Bootstrapping is a resampling technique to estimate statistics and confidence intervals by drawing samples (with replacement) from the data.

**Code:**


```python
import numpy as np

# Bootstrap sampling
population = df["population_total"].dropna()
bootstrap_samples = [
    np.mean(np.random.choice(population, size=len(population), replace=True))
    for _ in range(1000)
]

# Confidence Interval
ci_lower = np.percentile(bootstrap_samples, 2.5)
ci_upper = np.percentile(bootstrap_samples, 97.5)

print(f"95% Confidence Interval for Population Mean: [{ci_lower}, {ci_upper}]")
```


##### **A/B Testing**
**Theory:**
A/B Testing compares two groups (A and B) to evaluate the effect of a change (e.g., comparing two versions of a website for click rates).

**Code:**


```python
# Example: Comparing average literacy rates between two states
groupA = df[df["state_name"] == "PUNJAB"]["effective_literacy_rate_total"]
groupB = df[df["state_name"] == "MAHARASHTRA"]["effective_literacy_rate_total"]

# Perform t-test
t_stat, p_val = ttest_ind(groupA, groupB, nan_policy="omit")

print("T-Statistic for A/B Test:", t_stat)
print("P-Value for A/B Test:", p_val)
if p_val < 0.05:
    print("Significant difference in literacy rates between A and B.")
else:
    print("No significant difference in literacy rates between A and B.")
```

##### **Confidence Intervals for Accuracy**
**Theory:**
Confidence intervals indicate the range in which the true accuracy of a model lies with a certain probability.

**Code:**


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Example dataset preparation
df_cleaned = df.dropna()  # Ensure no missing values for simplicity
X = df_cleaned[["population_total", "effective_literacy_rate_total"]]
y = (df_cleaned["sex_ratio"] > 900).astype(int)  # Example target variable

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train a classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Calculate accuracy
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

# Bootstrapping for confidence interval
boot_accuracies = []
for _ in range(1000):
    sample_indices = np.random.choice(len(X_test), len(X_test), replace=True)
    boot_preds = model.predict(X_test.iloc[sample_indices])
    boot_acc = accuracy_score(y_test.iloc[sample_indices], boot_preds)
    boot_accuracies.append(boot_acc)

ci_lower = np.percentile(boot_accuracies, 2.5)
ci_upper = np.percentile(boot_accuracies, 97.5)

print(f"Accuracy: {accuracy:.2f}")
print(f"95% Confidence Interval for Accuracy: [{ci_lower:.2f}, {ci_upper:.2f}]")
```


##### **Comparing Classifiers**
**Theory:**
Classifiers can be compared using metrics like accuracy, precision, recall, and F1-score.

**Code:**


```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Train two classifiers
log_reg = LogisticRegression(random_state=42)
rf = RandomForestClassifier(random_state=42)

log_reg.fit(X_train, y_train)
rf.fit(X_train, y_train)

# Predictions
log_preds = log_reg.predict(X_test)
rf_preds = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report

# Comparison
print("Logistic Regression Performance:")
report_LR = classification_report(y_test, log_preds, output_dict=True)
# Convert the classification report to a DataFrame
report_LR = pd.DataFrame(report_LR).transpose()
# Display the DataFrame
display(report_LR)

print("Random Forest Performance:")
report_RF = classification_report(y_test, rf_preds, output_dict=True)
# Convert the classification report to a DataFrame
report_RF = pd.DataFrame(report_RF).transpose()
# Display the DataFrame
display(report_RF)
```
