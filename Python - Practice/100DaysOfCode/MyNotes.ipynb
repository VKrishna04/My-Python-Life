This python code converts a number from one base to another

````python
# Function to convert a number from one base to another
def convert_base(number, base1, base2):
  # Convert the number to base 10
  base10 = int(number, base1)
  
  # Convert the base 10 number to the target base
  target_number = ""
  while base10 > 0:
  remainder = base10 % base2
  target_number = str(remainder) + target_number
  base10 = base10 // base2
  
  return target_number

# Main program
number = input("Enter the number: ")
base1 = int(input("Enter the base of the number: "))
base2 = int(input("Enter the target base: "))

# Convert the number
converted_number = convert_base(number, base1, base2)

# Print the result
print("Converted number:", converted_number)
````
